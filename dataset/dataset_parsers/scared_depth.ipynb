```python
from PIL import Image
import numpy as np

import cv2
import numpy as np

def load_depth_with_opencv(filepath):
    """Load depth map using OpenCV"""
    # Load image with all channels
    img = cv2.imread(filepath, cv2.IMREAD_UNCHANGED)
    
    if img is None:
        raise ValueError("Could not load image")
    
    # OpenCV loads as BGR, so third channel is index 2
    if len(img.shape) == 3 and img.shape[2] >= 3:
        depth_map = img[:, :, 2]
    else:
        raise ValueError("Image doesn't have a third channel")
    
    return depth_map
# Usage
depth_map = load_depth_with_opencv('left_depth_map.tiff')
print(f"Depth map shape: {depth_map.shape}")
print(f"Depth range: {depth_map.min():.2f} - {depth_map.max():.2f} mm")
```
